```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

#Read in data
df = pd.read_csv(r"C:\Users\Kingsley Mills\Desktop\pretech\Analysis Projects\Analyst Portfolio Projects\movies.csv")
```

```python
# We look at the data 
df.head()
```

```python
# Lets check if there are any missing 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%' .format(col, pct_missing))
```

```python
#Let's remove missing values
df.dropna(inplace=True)
df.head()
```

```python
# Data types fro our coulumns
df.dtypes
```

```python
# Chamge data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
```

```python
# scatterplot with budget vs gross
x = df['budget']
y = df['gross']

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Firm')

plt.scatter(x,y)
plt.show()
```

```python
# Plot Budget vs Gross using Seaborn

sns.regplot(x,y, data = df, scatter_kws= {'color':'red'},line_kws ={'color':'blue'})
```

```python
# Lets start looking at correlation

df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Lets take a look at company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```
