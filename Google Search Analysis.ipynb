# Google Search Analysis

```python
# Import the necessary libraries
import pandas as pd  # Import pandas for data manipulation
from pytrends.request import TrendReq  # Import TrendReq from pytrends to fetch Google Trends data
import matplotlib.pyplot as plt  # Import matplotlib for plotting

# Create a TrendReq object to interact with Google Trends
trends = TrendReq()

# Now you can use the 'trends' object to make requests to Google Trends and fetch data.
```

```python
# Build a payload for Google Trends with the specified keyword(s)
# I used Coalition Technologies because I  applied for a Job
trends.build_payload(kw_list=["Coalition Technologies"])

# Request and fetch interest by region data for the specified keyword(s)
data = trends.interest_by_region()

# Sort the data by the "Coalition Technologies" column in descending order
data = data.sort_values(by="Coalition Technologies", ascending=False)

# Select the top 10 regions with the highest interest in "Coalition Technologies"
data = data.head(10)

# Print the resulting data, which represents the top 10 regions with the highest interest
print(data)
```

```python
# Reset the index of the DataFrame to make "geoName" a regular column
data.reset_index().plot(x="geoName", 
                        y="Coalition Technologies", 
                        figsize=(15, 12), kind="bar")

# Apply a specific style to the plot (in this case, the "fivethirtyeight" style)
plt.style.use('fivethirtyeight')

# Display the bar chart
plt.show()
```

```python
# Create a TrendReq object with specific settings (language and timezone)
data = TrendReq(hl='en-US', tz=360)

# Build a payload for Google Trends with the specified keyword
data.build_payload(kw_list=['Coalition Technologies'])

# Request and fetch interest over time data for the specified keyword
data = data.interest_over_time()

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(15, 12))

# Plot the data for "Coalition Technologies" over time
data['Coalition Technologies'].plot()

# Apply the "fivethirtyeight" style to the plot
plt.style.use('fivethirtyeight')

# Set the title of the plot
plt.title('Total Google Searches for Coalition Technologies', fontweight='bold')

# Set labels for the x and y axes
plt.xlabel('Year')
plt.ylabel('Total Count')

# Display the plot
plt.show()
```
